```python
from vdna import VDNAProcessor, EMD, load_vdna_from_files
import torch
import time
import matplotlib.pyplot as plt
import os
```

```python
vdna_proc = VDNAProcessor()
device = torch.device('cpu')
device = "cuda:0" if torch.cuda.is_available() else "cpu"

source1 = "source1/"
source2 = "source2/"

vdna1 = vdna_proc.make_vdna(source=source1, device=device, num_workers=0)
results = []
```

```python
source1 = "source1/"
source2 = "source2/"
```

```python
results = []
for image in os.listdir(source2):
    try:
        vdna2 = vdna_proc.make_vdna(source=source2 + '/' +image, device=device, num_workers=0)
        emd = EMD(vdna1, vdna2)
        data = {"image": image, "emd": emd} 
        results.append(data)
    except Exception as e:
        print(f"Error: {e}")
```

```python
results = sorted(results, key=lambda x: x['emd'])
```

```python
count = 0
for i in results:
    count += 1
    print(i['image'], i['emd'])
    
```
